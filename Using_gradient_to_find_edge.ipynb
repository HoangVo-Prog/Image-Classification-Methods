```python
image_path = '/content/cat.jpg'

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread(image_path, 0)

cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def computeXgradient(image):
  height, weight = image.shape
  x_gradient = np.zeros((height, weight))

  for i in range(1, height-1):
    for j in range(1, weight-1):
      x_gradient[i, j] = image[i, j+1] - image[i, j-1]

  return x_gradient

def computeYgradient(image):
  height, weight = image.shape
  y_gradient = np.zeros((height, weight))

  for i in range(1, height-1):
    for j in range(1, weight-1):
      y_gradient[i, j] = image[i+1, j] - image[i-1, j]

  return y_gradient


def computeMagnitude(x_gradient, y_gradient, image):
  height, weight = image.shape
  magnitude = np.zeros((height, weight))

  for i in range(1, height-1):
    for j in  range(1, weight-1):
      magnitude[i, j] = np.sqrt(x_gradient[i, j]**2 + y_gradient[i, j]**2)

  return magnitude


x_gradient = computeXgradient(image)
y_gradient = computeYgradient(image)
magnitude = computeMagnitude(x_gradient, y_gradient, image)

cv2_imshow(magnitude)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
